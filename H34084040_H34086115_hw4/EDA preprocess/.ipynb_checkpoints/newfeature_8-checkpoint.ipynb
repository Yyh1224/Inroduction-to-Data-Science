### Feature engineering

```python
train_new = train_o
train_new['TenureByAge'] = train_o.Tenure/(train_o.Age)
sns.boxplot(y='TenureByAge',x = 'Exited', hue = 'Exited',data = train_new)
test_new = test_o
test_new['TenureByAge'] = test_o.Tenure/(test_o.Age)
```

```python
train_new['BalanceSalaryRatio'] = train_new.Balance/train_new.EstimatedSalary
sns.boxplot(y='BalanceSalaryRatio',x = 'Exited', hue = 'Exited',data = train_new)
test_new['BalanceSalaryRatio'] = test_new.Balance/test_new.EstimatedSalary
```

```python
train_new['CreditScoreGivenAge'] = train_new.CreditScore/(train_new.Age)
test_new['CreditScoreGivenAge'] = test_new.CreditScore/(test_new.Age)
```

### test

```python
train_new['BalanceAge'] = train_new.Balance/(train_new.Age)
test_new['BalanceAge'] = test_new.Balance/(test_new.Age)
train_new['EstimatedSalaryAge'] = train_new.EstimatedSalary/(train_new.Age)
test_new['EstimatedSalaryAge'] = test_new.EstimatedSalary/(test_new.Age)
train_new['CreditScoreSalaryRatio'] = train_new.CreditScore/train_new.EstimatedSalary
test_new['CreditScoreSalaryRatio'] = test_new.CreditScore/test_new.EstimatedSalary
###
train_new['TenureSalaryRatio'] = train_new.Tenure/train_new.EstimatedSalary
test_new['TenureSalaryRatio'] = test_new.Tenure/test_new.EstimatedSalary
train_new['NumOfProductsSalaryRatio'] = train_new.NumOfProducts/train_new.EstimatedSalary
test_new['NumOfProductsSalaryRatio'] = test_new.NumOfProducts/test_new.EstimatedSalary
```
