### Data Analysis

###  replaced the continuous distribution into discrete values

```python
#distribution of credit score
plt.figure(figsize=(15,5))
sns.distplot(x=data["CreditScore"],axlabel="Credit Scores")
plt.show()
```

```python
df['CreditScoreBand'] = pd.cut(df['CreditScore'], 5)
df[['CreditScoreBand']].groupby(['CreditScoreBand'], as_index=False).mean().sort_values(by='CreditScoreBand', ascending=True)
```

```python
df_ = [df] #making a list so that 'for loops' can be easily implemented
```

```python
for dataset in df_:    
    dataset.loc[(dataset['CreditScore'] > 445.693) & (dataset['CreditScore'] <= 507.4), 'CreditScore'] = 0
    dataset.loc[(dataset['CreditScore'] > 507.4) & (dataset['CreditScore'] <= 568.8), 'CreditScore'] = 1
    dataset.loc[(dataset['CreditScore'] > 568.8) & (dataset['CreditScore'] <=  630.2), 'CreditScore'] = 2
    dataset.loc[(dataset['CreditScore'] > 630.2) & (dataset['CreditScore'] <= 691.6), 'CreditScore'] = 3
    dataset.loc[(dataset['CreditScore'] > 691.6), 'CreditScore'] = 4
```

```python
sns.countplot(x='CreditScore',data=df, color='blue', saturation=0.3)  
```

```python
df = df.drop(['CreditScoreBand'], axis=1)
df_ = [df]
df.head()
```

```python
#one-hot encoding categorical variables
lst = ['Geography', 'Gender']
remove = list()
for i in lst:
    if (df[i].dtype == np.str or df[i].dtype == np.object):
        for j in df[i].unique():
            df[i+'_'+j] = np.where(df[i] == j,1,-1)
        remove.append(i)
df = df.drop(remove, axis=1)
```

```python
#distribution of age
plt.figure(figsize=(10,5))
sns.distplot(x=data["Age"], bins=30, axlabel="Age")
plt.show()
```

```python
df['AgeBand'] = pd.cut(df['Age'], 5)
df[['AgeBand']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
df_ = [df]

for dataset in df_:    
    dataset.loc[(dataset['Age'] > 27.947) & (dataset['Age'] <= 38.6), 'Age'] = 0
    dataset.loc[(dataset['Age'] > 38.6) & (dataset['Age'] <= 49.2), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 49.2) & (dataset['Age'] <= 59.8), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 59.8) & (dataset['Age'] <= 70.4), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 70.4, 'Age'] = 4
df.head()
df = df.drop(['AgeBand'], axis=1)
df.head()
```

```python
sns.countplot(x='Age',data=df, color='blue', saturation=0.3)  
```
