### (1) Kmeans
1. 決定組數K，隨機選K個點做群集中心
2. 每個點分類到離自己最近的中心
3. 重新計算各組的群集中心
(2、3重複到中心不變)

Randomly select data points as the initialized centroids. By default, please set K=4. Report and 
plot  the  process  until  convergence.  The  centroids  also  need  to  be  plotted.  An  example  is 
shown below. Note that it may not have 3 rounds (it can be 4 or 5 rounds, depend on initialized 
centroids).

```python
def cal_dist(center, data):#calculate distance between center and data
    return np.sqrt(np.sum((center - data)**2, axis=1))
class KMeans:
    def __init__(self, n_clusters=5, round_count=100):
        self.n_clusters = n_clusters
        self.round_count = round_count
    def fit(self, train):
        #randomly select initial centroid
        self.centroids = [random.choice(train)]
        for i in range(self.n_clusters-1):
            #distance from centroid to data points
            for centroid in self.centroids:
                dist = np.sum([cal_dist(centroid, train)], axis = 0)
            # Normalize distances
            dist /= np.sum(dist)
            # Choose new centroid based on distances
            new_idx, = np.random.choice(range(len(train)), size=1, p=dist)
            self.centroids += [train[new_idx]]
        iter_count = 0
        prev_centroids = None
        #check if centroids list equal and iteration round enough
        while np.not_equal(self.centroids, prev_centroids).any() and iter_count < self.round_count:
            sorted_data = [[] for _ in range(self.n_clusters)]
            for x in train:
                dist = cal_dist(x, self.centroids)
                idx = np.argmin(dist)
                sorted_data[idx].append(x)
            # Push current centroids to previous, reassign centroids as mean of the points belonging to them
            prev_centroids = self.centroids
            self.centroids = [np.mean(cluster, axis=0) for cluster in sorted_data]
            for k, centroid in enumerate(self.centroids):
                if np.isnan(centroid).any():  # np.nans, resulting from a centroid having no points
                    self.centroids[k] = prev_centroids[k]
            iter_count += 1
    def evaluate(self, X):
        centroids = []
        idxs = []
        #return cluster and indexs
        for x in X:
            dists = cal_dist(x, self.centroids)
            centroid_idx = np.argmin(dists)
            centroids.append(self.centroids[centroid_idx])
            idxs.append(centroid_idx)
        return centroids, idxs
```

```python
def kmeans_sse(rd, centers_n = 4):#get the cluster for each data
    kmeans = KMeans(n_clusters=centers_n,round_count=rd)
    kmeans.fit(x_y)
    class_centers, classification = kmeans.evaluate(x_y)
    df_xy = pd.DataFrame(x_y, columns = ['x','y'])
    df_xy['cluster'] = pd.DataFrame(classification)
    return class_centers, classification,df_xy,kmeans
```

```python
def kmeans_round(rd_count, centers = 4):#with parameter of round 
    rd_count, centers, df_xy,kmeans = kmeans_sse(rd_count,4)
    marker_colors={0:'red',1:'yellow',2:'green',3:'blue'}
    plt.figure(figsize=(16,7))
    for i in range(4):#plot the scatter and clusters
        plt.plot(df_xy[df_xy['cluster']==i]['x'], 
                 df_xy[df_xy['cluster']==i]['y'],
                 linestyle='None', marker='o', markersize=10, 
                 markerfacecolor=marker_colors[i], label="cluster"+str(i))
    plt.plot([x for x, y in kmeans.centroids],
             [y for x, y in kmeans.centroids],
             'k+',
             markersize=10)
    plt.legend()
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
    return rd_count, centers
```

```python
for i in range(5):
    kmeans_round(i)
```

### (2)
- Re‐execute your K‐means clustering algorithm by changing K from 2 to 50 (from 2 to 10 is also okay). Plot the K value (x‐axis) vs. the value of Sum of Squared Error (SSE) (y‐axis) as below. Note that it is reasonable and acceptable if the curve is  凹凸不平. 

```python
sse_df = pd. DataFrame(columns=['k','sse_sum'])
df_xy = pd.DataFrame(x_y, columns = ['x','y'])
for i in range(2,51):
    #get centroid and data points for sse
    center_coor, cluster,df_xy,kmeans = kmeans_sse(4,i)
    temp = pd.DataFrame(center_coor, columns = ['centroid_x','centroid_y'])
    df_xy[['centroid_x','centroid_y']]= temp
    df_xy['cluster'] = pd.DataFrame(cluster)
    #calculate sse
    df_xy['sse'] = (df_xy['x']-df_xy['centroid_x'])**2 + (df_xy['y']-df_xy['centroid_y'])**2
    #sum sse
    total = df_xy['sse'].sum()
    sse_df = sse_df.append({'k': i , 'sse_sum':total},ignore_index=True)
```

```python
#plot sse
plt.plot(sse_df['k'], sse_df['sse_sum'], color='blue')
plt.xlabel("k")
plt.ylabel("sse")
```

### (3)
- Try 10 times of randomly initialized centroids, and plot their SSE values (y‐axis) 


```python
sse_rand = pd.DataFrame(columns=['time','Sum of square error'])
```

```python
for l in range(10):
    # randomly pick 3 initial points
    centroids_k = [random.choices(x_y, k = 10)]
    #process to df
    lst1 = []
    lst2 = []
    for i in range(10):
        lst1.append(centroids_k[0][i][0])
        lst2.append(centroids_k[0][i][1])
    df_centroids = pd.DataFrame(list(zip(lst1, lst2)), columns =['centroid_x','centroid_y']) 
    df_rand = df_xy[['x','y']]
    ls_idx = []
    #classify data
    for j in range(300):
        data_arr = np.array(df_rand.loc[j])
        centroid_arr = np.array(df_centroids)
        dists = cal_dist(data_arr, df_centroids)
        centroid_idx = np.argmin(dists) 
        ls_idx.append(centroid_idx)
    df_rand['cluster'] = ls_idx
    df_merge_rand = pd.merge(df_rand, df_centroids, left_on=["cluster"], right_index=True)
    #calculate sse
    df_merge_rand['sse'] = (df_merge_rand['x']-df_merge_rand['centroid_x'])**2 + (df_merge_rand['y']-df_merge_rand['centroid_y'])**2
    sse_rand= sse_rand.append({'time': l , 'Sum of square error':df_merge_rand['sse'].sum()},ignore_index=True)
```

```python
#plot sse
plt.bar(sse_rand['time'], sse_rand['Sum of square error'], color="blue")
plt.xlabel('Fixed k = 10') 
plt.ylabel('Sum of square error') 
  
# displaying the title
plt.title("Result of ten times randomly pick the initial points")
```
